```python
import pandas as pd
df = pd.read_csv("Summer-Olympic-medals-1976-to-2008.csv") 
df.head()
```

```python
df=df.fillna(0)
df['Year']= df['Year'].astype('int')
df.set_index('Year', inplace=True)
df.head()
```

```python
name = df['Athlete'].str.split(',',expand=True,n=1)
name.columns = ['Athlete_Last_name', 'Athlete_First_name']
df = pd.concat([df, name], axis=1)
df.drop(['Athlete'],axis=1,inplace=True)
df.head()

```

```python
freq = df.groupby('Country_Code').agg({'Medal':'count'}) 
freq=freq.sort_values('Medal',ascending=False)
print(freq)
```

```python

print("El pais que más medallas obtuvo fue " + freq.iloc[0].name + " que obtuvo:" +  str(freq.iloc[0][0]) + " Medallas" )
#print(freq.index)
```
