```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("HRDataset_v14.csv") 

df.head()
```

```python
df=df.fillna(0)
df.dtypes
```

```python
df.describe(include='all')
```

```python

df['GenderID'].replace(0, 'Female',inplace=True)
df['GenderID'].replace(1, 'Male',inplace=True)
sns.barplot(x="GenderID", y=df.index,data=df).set_title('Cantidad de empleados mujeres y hombres')
```

```python
sns.barplot(y="RecruitmentSource", x=df.index,data=df)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.hist(df['Salary'], bins=30, density=True)
ax.set_xlabel('Ingreso [$]') 
ax.set_title('Distribucion de salarios') 
```

Vamos que el rango de salarios más popular es de 50000 a 60000  

```python
fig = plt.figure()  
#sns.set(rc={'figure.figsize':(15,15)})
sns.violinplot(x=df['GenderID'], y=df['Salary'], data=df)
plt.show()
```

```python
print(df['DateofHire'])
```

```python
df_graficar=pd.DataFrame()
df_graficar['NumerHired']=pd.to_datetime(df['DateofHire']).dt.year.value_counts()
df_graficar=df_graficar.sort_index()
print(df_graficar)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(df_graficar.index,df_graficar['NumerHired'])
ax.set_ylabel('Cantidad de empeados contratados') 
ax.set_xlabel('Año de contratación') 
ax.set_title('Empleados contratados por año') 


```
