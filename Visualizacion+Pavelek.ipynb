```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("HRDataset_v14.csv") 

df.head()
```

```python
df['GenderID'].replace(0, 'Female',inplace=True)
df['GenderID'].replace(1, 'Male',inplace=True)


sns.barplot(x="GenderID", y=df.index,data=df)

```

```python
sns.barplot(y="RecruitmentSource", x=df.index,data=df)
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.hist(df['Salary'], bins=10, density=True)
ax.set_xlabel('Ingreso [$]') 
ax.set_title('Distribucion de salarios') 
```

Vamos que el rango de salarios más popular es de /$50000 a /$66000

```python
fig = plt.figure()  
sns.set(rc={'figure.figsize':(15,15)})
sns.violinplot(x=df['GenderID'], y=df['Salary'], data=df)
plt.show()
```

```python
print(df['DateofHire'])
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(df.index ,df['DateofHire'], label='Año de contratación')

ax.set_xlabel('Empleado') 
ax.set_ylabel('Año de contratación') 
ax.set_title('Lluvias de los Eneros') 
ax.legend()

ax.set_xlim(df.index[0], df.index[-1] )
```
